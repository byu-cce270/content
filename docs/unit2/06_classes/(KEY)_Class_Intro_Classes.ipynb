# Exercise 1 - Student Class

In the first exercise, we will practice writing the pattern for a class in different steps.

## Step 1

Below in Code Block 1, you will define the class 'Student'. The class should have the following attributes:

- first name
- last name
- major
- year, the default year should be "Freshman"
- gpa, you should be able to not put a GPA in. Also set it up so any GPA scores entered are validated and between 0 and 4 (the GPA can also be 0 or 4). Print a message if a score entered does not follow this validation and then set the score to None.

```python
# @title Code Block 1

class Student:
    def __init__(self, first_name, last_name, major, year="Freshman", gpa=None):
        self.first_name = first_name
        self.last_name = last_name
        self.major = major
        self.year = year
        if gpa is not None and (gpa < 0 or gpa > 4):
            print(f"Invalid GPA: {gpa}. GPA must be between 0 and 4. Setting GPA to None.")
            self.gpa = None
        else:
            self.gpa = gpa
```

## Step 3

In Code Block 3, create a new class called 'Full_Student' by copying your 'describe_Student' class from Code Block 2.

Add two more methods: 'update_year' and 'is_honors'.

'update_year' should update the student's year.
'is_honors' should check to see if the student's GPA is greater than or equal to 3.5. If the student is or is not an honors student, print a message saying so.

```python
# @title Code Block 3

class Full_Student:
    def __init__(self, first_name, last_name, major, year="Freshman", gpa=None):
        self.first_name = first_name
        self.last_name = last_name
        self.major = major
        self.year = year
        if gpa is not None and (gpa < 0 or gpa > 4):
            print(f"Invalid GPA: {gpa}. GPA must be between 0 and 4. Setting GPA to None.")
            self.gpa = None
        else:
            self.gpa = gpa

    def describe(self):
        print(f"{self.first_name} {self.last_name} is a {self.year} majoring in {self.major}.")

    def update_year(self, new_year):
        self.year = new_year

    def is_honors(self):
        if self.gpa is not None and self.gpa >= 3.5:
            print(f"{self.first_name} {self.last_name} is an honors student with a GPA of {self.gpa}.")
        else:
            print(f"{self.first_name} {self.last_name} is not an honors student.")
```

## Vehicle Class

In Code Block 4, create `Vehicle`:
- `make`, `model`, `year`, `odometer` (start at 0)
- `drive(miles)` adds miles to odometer

Also, create `ConstructionVehicle` (inherits from `Vehicle`):
- add `equipment_type` and `engine_hours` (start at 0)
- `run(hours)` adds to engine_hours
- `needs_service()` if miles or hours exceed a threshold you choose

```python
# @title Code Block 4

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer = 0

    def drive(self, miles):
        self.odometer += miles


class ConstructionVehicle(Vehicle):
    def __init__(self, make, model, year, equipment_type):
        super().__init__(make, model, year)
        self.equipment_type = equipment_type
        self.engine_hours = 0

    def run(self, hours):
        self.engine_hours += hours

    def needs_service(self):
        if self.odometer > 50000 or self.engine_hours > 500:
            print(f"{self.make} {self.model} ({self.equipment_type}) needs service!")
            return True
        else:
            print(f"{self.make} {self.model} ({self.equipment_type}) does not need service yet.")
            return False
```

Test your 'ConstructionVehicle' class below.

```python
# @title Code Block 4 Test

excavator = ConstructionVehicle("Caterpillar", "320", 2022, "Excavator")
excavator.drive(45000)
excavator.run(480)
print(f"Odometer: {excavator.odometer} miles")
print(f"Engine hours: {excavator.engine_hours} hours")
excavator.needs_service()

print()

# Add more miles to trigger service
excavator.drive(6000)
excavator.needs_service()
```

## ConcreteBatch Class

In Code Block 5, create `ConcreteBatch` with:
- `cement_kg`, `water_kg`, `aggregate_kg`
- `water_cement_ratio()` returns water_kg / cement_kg
- `total_mass()` returns sum
- `summary()` prints the mix + w/c ratio
- validate: cement can't be 0; all masses > 0
- `meets_wc_limit(limit)` (example limit = 0.50)

```python
# @title Code Block 5

class ConcreteBatch:
    def __init__(self, cement_kg, water_kg, aggregate_kg):
        if cement_kg <= 0:
            print("Error: Cement mass must be greater than 0.")
            cement_kg = 1
        if water_kg <= 0:
            print("Error: Water mass must be greater than 0.")
            water_kg = 1
        if aggregate_kg <= 0:
            print("Error: Aggregate mass must be greater than 0.")
            aggregate_kg = 1
        self.cement_kg = cement_kg
        self.water_kg = water_kg
        self.aggregate_kg = aggregate_kg

    def water_cement_ratio(self):
        return self.water_kg / self.cement_kg

    def total_mass(self):
        return self.cement_kg + self.water_kg + self.aggregate_kg

    def summary(self):
        print(f"Concrete Mix: {self.cement_kg} kg cement, {self.water_kg} kg water, {self.aggregate_kg} kg aggregate")
        print(f"Total mass: {self.total_mass()} kg")
        print(f"Water/cement ratio: {self.water_cement_ratio():.2f}")

    def meets_wc_limit(self, limit=0.50):
        wc = self.water_cement_ratio()
        if wc <= limit:
            print(f"W/C ratio of {wc:.2f} meets the limit of {limit}.")
            return True
        else:
            print(f"W/C ratio of {wc:.2f} exceeds the limit of {limit}.")
            return False
```

Test your 'ConcreteBatch' class below.

```python
# @title Code Block 5 Test

batch1 = ConcreteBatch(350, 160, 1200)
batch1.summary()
batch1.meets_wc_limit(0.50)

print()

batch2 = ConcreteBatch(300, 180, 1100)
batch2.summary()
batch2.meets_wc_limit(0.50)

print()

# Test validation
batch3 = ConcreteBatch(0, 100, 500)
```

# Bonus Exercise - BankAccount

If time permits or you still want more practice, create `BankAccount` in code block 6 with:
- `owner`, `balance` (default 0)
- `deposit(amount)`
- `withdraw(amount)` (prevent overdraft: print a message)
- `transactions` list of transactions: type, amount, balance
- `transaction_history()` (prints list of transactions

```python
# @title Code Block 6

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transactions.append(f"Deposit: +${amount:.2f}, Balance: ${self.balance:.2f}")
            print(f"Deposited ${amount:.2f}. Current balance: ${self.balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient funds. Current balance: ${self.balance:.2f}")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.balance -= amount
            self.transactions.append(f"Withdrawal: -${amount:.2f}, Balance: ${self.balance:.2f}")
            print(f"Withdrew ${amount:.2f}. Current balance: ${self.balance:.2f}")

    def transaction_history(self):
        print("\nTransaction history:")
        for t in self.transactions:
            print(t)
```

Test your 'BankAccount' class below.

```python
# @title Code Block 6 Test

account = BankAccount("John Smith", 1000)
account.deposit(500)
account.withdraw(200)
account.withdraw(2000)  # Should print insufficient funds

account.transaction_history()
```
