```python
# 3-6. More Guests: You just found a bigger dinner table, so now more space is available. Think of three more guests to invite to dinner.
# Start with your program from Exercise 3-4 or 3-5. Add a print() call to the end of your program, informing people that you found a bigger table.
# Use insert() to add one new guest to the beginning of your list.
# Use insert() to add one new guest to the middle of your list.
# Use append() to add one new guest to the end of your list.
# Print a new set of invitation messages, one for each person in your list.

guest_list = ["Norm Jones","Gus Williams","Cosmo the Cougar"]

print(f"Hello {guest_list[0]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[1]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[2]}, you are formally invited to dinner at Olive Garden!")

print(f"\nUnfortunately, {guest_list[2]} cannot make it to dinner.")

guest_list[2] = "Kalani Sitake"
print(f"\nHello {guest_list[0]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[1]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[2]}, you are formally invited to dinner at Olive Garden!")

print("\nGreat news everybody! We found a bigger table, so we will invite 3 more guests!")

guest_list.insert(0,"C. Shane Reese")
guest_list.insert(2,"Bear Bachmeier")
guest_list.append("Tiger Bachmeier")

print(f"\nHello {guest_list[0]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[1]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[2]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[3]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[4]}, you are formally invited to dinner at Olive Garden!")
print(f"Hello {guest_list[5]}, you are formally invited to dinner at Olive Garden!")
```
